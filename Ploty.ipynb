```python
import seaborn as sns
import plotly.express as plt
data=sns.load_dataset("titanic")

scat = plt.scatter(data , x="age",y="fare")
scat.show()
```

# ANswer 2

```python
# Load the tips dataset from Plotly
tips_data = plt.data.tips()

# Create a box plot for total bill and day of the week
box_plot = plt.box(tips_data, x="day", y="total_bill")

# Show the box plot
box_plot.show()
```

# Answer 3

```python
# Load the tips dataset from Plotly
tips_data = plt.data.tips()

# Create a histogram for total bill, with sex on the x-axis and smoker on the pattern_shape parameter
histogram = plt.histogram(tips_data, x="sex", y="total_bill", color="day", pattern_shape="smoker")

# Show the histogram
histogram.show()
```

# Answer 4

```python
iris_data = plt.data.iris()

# Create a scatter matrix plot
scatter_matrix = plt.scatter_matrix(
    iris_data,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species"
)

# Show the scatter matrix plot
scatter_matrix.show()
```
