```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.title("This is title")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")

plt.scatter(x,y)
plt.show()
```

# subplot() 
function adds subplot to a current figure at the specified grid position. It is similar to the subplots() function however unlike subplots() it adds one subplot at a time

# plot 1

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(1,2,1)

plt.plot(x,y)
plt.show()
```

# barplot
(or barchart) is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. 

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
```

# ans5 

Box plots provide a quick visual summary of the variability of values in a dataset. They show the median, upper and lower quartiles, minimum and maximum values, and any outliers in the dataset. Outliers can reveal mistakes or unusual occurrences in data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
fig = plt.figure(figsize =(10, 7))
 

ax = fig.add_axes([0, 0, 1, 1])
 
bp = ax.boxplot(data)
plt.show()
```
