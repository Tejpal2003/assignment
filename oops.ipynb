```python
Skip to content
 
Search…
All gists
Back to GitHub
@Tejpal2003 
@AvinashGawai
AvinashGawai/assignment loops.py
Created 6 months ago • Report abuse
Code
Revisions
1
<script src="https://gist.github.com/AvinashGawai/057f376707b0bafa08fb79579299bf32.js"></script>
PWSKILLS DATA SCIENCE DS LOOPS ASSIGNMENT
assignment loops.py
# Q.1) Explain with an example each when to use a for loop and a while loop.
# For loop: A for loop is a control flow statement that executes code
# repeatedly for a particular number of iterations. In this control flow
# statement, the keyword used is for. The for loop is used when the
# number of iterations is already known.
# While loop: A "while" loop is used when the number of iterations is unknown.
# The statement repeats itself till the boolean value becomes false.
# Example 1) for loop
for i in range(0,10+1):
    print(i)
# # Example 2) while loop
n = int(input("Enter the number"))
i = 0
while i<n:
    i = i + 1
    print(i)

# # Q2. Write a python program to print the sum and product of the first
# # 10 natural numbers using for and while loop.
print("using for loop")
sum = 1
sum1 = 0
for i in range(1,11):
    sum = sum * i
    sum1 = sum1 + i
print(sum)
print(sum1)

print("using while loop")
i = 1
sum2 = 1
sum3 = 0
while i < 11:
    sum2 = sum2 * i
    sum3 = sum3 + i
    i+=1
print(sum2)
print(sum3)

# # # '''Electricity Program'''
# # Q3. Create a python program to compute the electricity bill for a household.
unit = int(input())
if (unit > 0 and unit <=100):
    print(100*4.5)
elif (unit > 100 and unit < 200):
    print((100*4.5)+(unit-100)*6)
elif (unit>200 and unit < 300):
    print((100*4.5)+(100*6)+(unit-300)*10)
elif (unit > 300):
    print((100*4.5)+(100*6)+(100*10)+(unit-300)*20)

# Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
# number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
# that list.
l1 = list(range(1,101))
l2 = []
l3 = []
l4 = []
l5 = []
print("1 to 100 List")
print(l1)
print("Using For Loop")
for i in l1:
    cube = i**3
    l2.append(cube)
print(l2)
for i in l2:
    if i % 4 == 0 or i % 5 == 0:
        l3.append(i)
print(l3)
print("Using While Loop")
i = 1
while i<len(l1)+1:
    ans = i**3
    l4.append(ans)
    i+=1
print(l4)
for i in l4:
    if i % 4 == 0 or i % 5 == 0:
        l5.append(i)
print(l5)
#
# # Q5. Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"
string.lower()
list = ['a','e','i','o','u']
l2 = []
for i in list:
    for j in string.lower():
            if i == j :
                l2.append(i)
print(l2)
```
